```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/content/US_Accidents_March23_sampled_500k.csv")
df
```

```python
df['Start_Time'] = pd.to_datetime(df['Start_Time']) # 'Start_Time' to datetime format
```

```python
df['Hour'] = df['Start_Time'].dt.hour #Extract hour from 'Start_Time'
```

```python
df.dtypes
```

```python
df.info()
```

```python
numerical_attributes = ['Severity', 'Start_Lat', 'Start_Lng', 'End_Lat', 'End_Lng', 'Distance(mi)', 'Temperature(F)']
df_numerical = df[numerical_attributes]
correlation_matrix = df_numerical.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Severity',color='red',data=df)
plt.title('Distribution of Accident Severity')
plt.xlabel('Severity')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Source', data=df)
plt.title('Distribution of Accident Sources')
plt.xlabel('Source')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Sunrise_Sunset',color='red',data=df)
plt.title('Distribution of Accidents by Time of Day')
plt.xlabel('Time of Day (Sunrise/Sunset)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Weather_Condition', data=df, order=df['Weather_Condition'].value_counts().index[:10])
plt.title('Top 10 Weather Conditions During Accidents')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Traffic_Signal',color = 'red',data=df)
plt.title('Distribution of Accidents by Traffic Signal')
plt.xlabel('Traffic Signal Availability')
plt.ylabel('Count')
plt.show()
```

```python
if 'Temperature(F)' in df.columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='Severity', y='Temperature(F)', data=df)
    plt.title('Box Plot of Temperature by Severity')
    plt.xlabel('Severity')
    plt.ylabel('Temperature (F)')
    plt.show()
```

```python
weather_counts = df['Weather_Condition'].value_counts().head(5)
plt.figure(figsize=(8, 8))
plt.pie(weather_counts, labels=weather_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Weather Condition Distribution (Top 5)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Severity', y='End_Lng',color='red', data=df)
plt.title('Box Plot of Severity vs. End Longitude')
plt.xlabel('Severity')
plt.ylabel('End Longitude')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Start_Lng', y='Start_Lat', data=df, alpha=0.1)
plt.title('Accident Hotspots')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
